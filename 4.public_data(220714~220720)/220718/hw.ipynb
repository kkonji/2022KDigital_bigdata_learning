```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import csv
import platform
%matplotlib inline
```

# 방법1

```python
def make(city):
    f = open('../DATA_WOO/gender.csv', encoding='euc-kr')
    data =csv.reader(f)
    male_sum = 0
    female_sum = 0
    for row in data:
        if city in row[0]:
            male_sum = int(row[104].replace(',',''))
            female_sum = int(row[104+103].replace(',',''))
            break
    f.close()
    if	platform.system()	==	'Windows':
        plt.rc('font',	family='Malgun Gothic')
    else:
        plt.rc('font',	family='AppleGothic')
    color = ['cornflowerblue', 'tomato']
    plt.pie([male_sum, female_sum], labels=['남성', '여성'], autopct='%.1f%%', colors=color, startangle=90)
    
    if ' ' in city:
        plt.title(f'{city}' )
    else:
        plt.title(f'{city} 전체')
   
    plt.legend()



region = ['대구광역시 중구', '대구광역시 동구', '대구광역시 서구', '대구광역시 남구', '대구광역시 달서구', '대구광역시 수성구', '대구광역시 북구', '대구광역시']
fig, axes = plt.subplots(2,4, figsize=(12,9), sharex=True, sharey=True)

fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=15)

for i in range(1,9):
    plt.subplot(2,4,i)
    make(region[i-1])
    
fig.tight_layout()
plt.show()
```
