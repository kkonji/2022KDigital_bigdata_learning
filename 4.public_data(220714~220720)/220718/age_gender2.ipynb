### barh(y, data) 그래프

```python
import platform
def print_population(pop_list):
    '''
    input: 인구 리스트 -> return: 인구 수 프린트

    '''
    for idx, val in enumerate(pop_list):
        print(f'{idx:3d}세 {val:4d}명', end=' ')
        if (idx+1) % 10 == 0:
            print()
def draw_population(man_list, female_list):
    if	platform.system()	==	'Windows':
        plt.rc('font',	family='Malgun Gothic')
    else:
        plt.rc('font',	family='AppleGothic')

    # barh 그래프
    plt.barh(range(len(man_list)), man_list, label= '남성')
    plt.barh(range(len(female_list)), female_list, label= '여성')
    plt.rcParams['axes.unicode_minus']	=	False
    plt.legend()
    plt.show()


def cal_population():
    f = open('../DATA_WOO/gender.csv', encoding='euc-kr')
    data =csv.reader(f)
    male_list = []
    female_list = []

    distinct = input('지역을 입력하세요: ')
    for row in data:
        if distinct in row[0]:
            for male in row[106:207]:
                if ',' in male:
                    male = male.replace(',','')
                male_list.append(-int(male))
            for female in row[209:310]:
                if ',' in female:
                    female = female.replace(',','')
                female_list.append(int(female))
    f.close()
    print(f'남성 총인구: ', sum(male_list))
    print_population(male_list)
    print(f'여성 총인구: ', sum(female_list))
    print_population(female_list)
    draw_population(male_list, female_list)



cal_population()
```

### 산점도 그래프

```python
import csv
import matplotlib.pyplot as plt
import platform
import math
def cal_population():
    f = open('../DATA_WOO/gender.csv', encoding='euc-kr')
    data =csv.reader(f)
    male_list = []
    female_list = []
    size_list = []

    city = input('지역을 입력하세요: ')
    for row in data:
        if city in row[0]:
            for i in range(106,207):
                male_num =int(row[i].replace(',',''))
                female_num = int(row[i+103].replace(',',''))
                male_list.append(male_num)
                female_list.append(female_num)
                size_list.append(math.sqrt(male_num + female_num))
            break
    f.close()
    draw_scatter(city, male_list, female_list, size_list)

def draw_scatter(city, male_list, female_list, size_list):
    if	platform.system()	==	'Windows':
        plt.rc('font',	family='Malgun Gothic')
    else:
        plt.rc('font',	family='AppleGothic')
    # color = ['cornflowerblue', 'tomato']
    plt.figure(figsize=(8,4), dpi=100)
  
    plt.scatter(male_list, female_list, s=size_list, c=range(101), alpha=0.5, cmap='rainbow')
    plt.colorbar()
    plt.plot(range(max(male_list)),range(max(male_list)) , 'g--')
    plt.title(f'{city} 지역별 남녀 인구수 비교' )
    plt.xlabel('남자인구수')
    plt.ylabel('여자인구수')
    plt.show()

cal_population()
```

### 정규식과 불린대수를 이용해서 특정조건 만족하는 row만 뽑기

```python
import re
word = '대구광역시 남구 대명동'
# 행정구역이 ~시 ~구 로 시작하는 row만 뽑고싶다.
def word_extract(word):
    # 행정구역: ~시
    si = re.compile(r'(\w+[시])')
    # 행정구역: ~구
    goo = re.compile(r'\w+\s(\w+[구])')
    si = re.match(si, word)
    # 매치가 성립하지 않으면 False를 리턴
    if not si:  return False
    goo = re.match(goo, word)
    # 매치가 성립하면 True를 리턴
    if goo: 
        return True
    else: return False
    

# word_extract('부산광역 사하구')
# 테스트
a = df['행정구역'][1203]
print(word_extract(a))
a = df['행정구역'][903]
print(word_extract(a))
a = df['행정구역'][2203]
print(word_extract(a))

# 함수를 적용해서 불린 리스트 값을 얻고 그걸 조건으로 하여 특정 row를 고른다. 
df.loc[df['행정구역'].apply(word_extract), :]
```
