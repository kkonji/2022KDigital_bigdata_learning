# DF 객체 연결하기

- 여러 개의 파일에서 읽은 df객체
- 메서드 실행 후 결과인 df객체
- 여러개의 df객체를 연결 시켜서 하나의 df객체 생성
- pandas.concat(데이터프레임 리스트, axis =0,join='inner'(or 'outer')

```python
import pandas as pd
```

```python
PATH = '../DATA/'
FILENAME1 = PATH + 'concat_1.csv'
FILENAME2 = PATH + 'concat_2.csv'
FILENAME3 = PATH + 'concat_3.csv'
df1 = pd.read_csv(FILENAME1)
df2 = pd.read_csv(FILENAME2)
df3 = pd.read_csv(FILENAME3)
print(df1)
print(df2)
print(df3)
```

```python
result1 = pd.concat([df1, df2])
result1
```

```python
result2 = pd.concat([df1,df2,df3])
result2
```

```python
result3 = pd.concat([df1,df2], join='inner')
result3
```

```python
df4 = df1[['A', 'B', 'C']].copy()
df4
```

```python
df5 = pd.concat([df2, df4])
df5
```

```python
df6 = pd.concat([df2, df4], join='inner')  #  join = 'inner'를 사용하면 공통인 열만 들어간다. 
df6
```

```python
# axis=1을 사용해서 열방향 연결
df7 = pd.concat([df2,df4], axis = 1, ignore_index=True)
df7
```

```python
# 행방향 연결, ignore_index를 사용해서 기본 인덱스 유지 x, 컬럼 교집합
df8 = pd.concat([df1,df2,df3], join='inner', ignore_index=True)
df8
```

```python
df0 = [['a0','b0'], ['a1','b1']]
df0 = pd.DataFrame(df0)
df0
```

```python
df10 = pd.concat([df0,df0], axis =1, ignore_index= True)
df10 = pd.concat([df10, df0], ignore_index=True)
df10 = pd.concat([df10, df0], axis = 1, ignore_index=True)
df10
```

```python
df = pd.concat([df1,df2,df3], keys=['DF1','DF2','DF3'] ) # keys를 넣으면 멀티인덱스로 넣을 수 있다. 
df
```
