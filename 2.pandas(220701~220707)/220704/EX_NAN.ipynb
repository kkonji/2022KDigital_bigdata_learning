# 결측치(Missing Value)처리 실습 

- 결측치(Missing Value)란?
    : 값이 들어있지 않은 것으로 NaN, NA로 표기
    
- 관련 메서드
    - 결측치 여부 체크: isnull(), isna(), notnull(), notna()
    - 결측치 삭제: dropna()
    - 결측치 치환: fillna()

```python
# 모듈 로딩
import pandas as pd
```

```python
# 파일 관련 변수 선언
PATH = '../DATA/'
FILENAME2 = PATH + 'weather.csv'
# csv file -> DataFrame
weatherDf = pd.read_csv(FILENAME2)
weatherDf
```

```python
# 데이터 정보 확인
weatherDf.info()
```

```python
weatherDf.shape
```

```python
weatherDf.describe().loc[:,'d1':'d10']
```

```python
# 결측치 개수 파악
weatherDf.isnull().sum()
```

```python
# 결측치 개수 파악
weatherDf.isnull().sum() == 22
```

```python
# 데이터 삭제
weatherDf2 = weatherDf.dropna(axis = 1, how = 'all' )  # 최고, 최저 온도값이 모두 NaN인 것 삭제
weatherDf2
```

```python
oneMonth = weatherDf2.iloc[0:2]
oneMonth = oneMonth.dropna(axis=1, how = 'all' )
oneMonth
```

```python
oneMonth = weatherDf2.iloc[0:8]
oneMonth = oneMonth.dropna(axis=1, how = 'all' )

monthMax = oneMonth[oneMonth['element'] == 'tmax']
monthMin = oneMonth[oneMonth['element'] == 'tmin']

monthMax = monthMax.iloc[:,4:].max(axis=1, numeric_only=True)
monthMin = monthMin.iloc[:,4:].min(axis=1, numeric_only=True)

Maxmin = pd.concat([monthMax, monthMin]).reindex(oneMonth.index)
print(type(Maxmin))
oneMonth['월별 최고,최저 기온'] = Maxmin
oneMonth
```
