```python
# 전처리 -< NaN, Duplicate, Datetype
# Date 컬럼 : object -> datetime64
# 방법1: 현재 데이터를 가져와서 타입을 바꾸고 반환(astype)
stockDf['Date'] = stockDf['Date'].astype('datetime64') 
# 방법2: 판다스의 to_datetime() 이용
stockDf['Date'] = stockDf['Date'].astype('datetime64') 

# # 인덱스 설정
# stockDf.set_index('Date', inplace=True)
```

```python
stockDf.head(5)
```

```python
# 특정 데이터 추출 6/20~25
day = stockDf['Date'].dt.day
month = stockDf['Date'].dt.month
cond = (20 <= day) & (day <= 25) & (month == 6)
stockDf[cond]
```

```python
# Index 변경 -> Date 컬럼을 인덱스 설정
stockDf.set_index('Date', inplace=True)
stockDf
```

```python
# 6/20부터 25일까지 데이터만 추출 -> df.loc[]
cond = (stockDf.index.day >= 20) & (stockDf.index.day <= 25)
stockDf.loc[cond]
```

```python
# 2018년도만 추출
cond = (stockDf.index.year == 2018)
stockDf.loc[cond]
```

```python
# 2018년도만 추출2
stockDf.loc['2018']
```

```python
# 데이터프레임에서 고유값 unique로 뽑아와서 리스트에 담기
def dataframe_unique(someDF):
    somelist = [col for col in someDF.columns]
    for col in somelist:
        print(f'{col}의 고유값: {someDF[col].unique()}')
    return somelist
dataframe_unique(stockDf)
```
