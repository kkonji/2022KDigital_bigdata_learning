# 접근자(Accessor) 실습
- DF/SR에 속성으로 제공되며, 다른객체의 기능, 속성을 사용할 수 있도록 해줌.
- 종류
    - Series.dt : datetime 관련 속성 사용할 수 있도록 해줌
    - Series.str : string 관련 메서드 사용할 수 있도록 해줌
    - DataFrame.spare : spare 관련 속성/메서드 사용할 수 있도록 해줌
    

```python
# 모듈 로딩
import pandas as pd 

# 데이터 생성

# 이름, 생년월일, 나이
df = pd.DataFrame({'name':['Tom kim', 'Red Park', 'Nale Min', 'Sezan Ho', 'Yellow Hello', 'Main Park', 'Dal Sep', 'Church Done', 'Lake Horse','Butter Fly'], 'age':[29, 29, 29, 29, 29, 29, 29, 29, 29, 29]})
df
```

```python
df['birth'] = pd.date_range(start='1/4/1994', end= '11/6/1994', freq= 'M') + pd.Timedelta(4, unit='D')
today = pd.Timestamp('2022-07-10') - df['birth']
age = today / 365
age
```

## str 데이터로 구성된 컬럼의 str 접근자
- 전체 컬럼의 요소에 str관련 메서드 사용 가능하도록 하는 접근자
- 사용법
    - 객체변수[컬럼명].str.메서드() -> upper, lower, split, replace, ...
    - 객체변수[컬럼명].str.get(인덱스) -> 분리 문자 리스트에서 특정 요소 추출
    
    

```python
# 문자열 분리
df[['first name', 'last name']] = df['name'].str.split(expand=True)

# 분리된 문자열 리스트에서 특정 값 가져오기
# df['name'].str.split().str.get(0), df['name'].str.split().str.get(1) 방법도 있다. 
df
```

```python

df2 = df['name'].str.split(expand=True)
pd.concat([df, df2], axis=1)
```

```python
# 여러가지 예시
df['name'].str.lower(), df['name'].str.upper()
```
