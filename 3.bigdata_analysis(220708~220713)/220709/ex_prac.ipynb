```python
import pandas as pd
PATH = '../DATA_WOO/'
df = pd.read_csv(PATH +'stock-data.csv')

df
```

```python
# 'Date'컬럼을 연 월 일로 쪼개서 넣기

df[['Y', 'M', 'D']] = df['Date'].str.split('-', expand=True)
df = df.reindex(columns= ['Date', 'Y', 'M', 'D', 'Close', 'Start', 'High', 'Low', 'Volume'])
df.sort_values(by='D', inplace=True)
```
