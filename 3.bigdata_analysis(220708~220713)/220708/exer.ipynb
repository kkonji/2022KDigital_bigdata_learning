```python
df.dtypes

def normalize(df, col):
    """
    df = DataFrame, col = column(int or float) -> Add normalized_col 
    방식 : (컬럼의 데이터- 평균)/ 표준편차 -> Z-score 
    """
    col_mean = df[col].mean()
    col_std = df[col].std()
    col_nbr = df.columns.get_loc(col)+1
    df.insert(col_nbr, f'{col}_normalize', 0)
    df[f'{col}_normalize'] = (df[col] - col_mean) / col_std

def catch_int_col(df):
    """
    df = DataFrame -> return: column 명(int, float)
    아직은 float64와 int64만 지원
    """
    somelist = [(df.dtypes[num] == 'float64') or (df.dtypes[num] == 'int64') for num in range(len(df.columns))]
    return df.loc[:,somelist].columns

someList = catch_int_col(df)
for col in someList:
    normalize(df, col)
```
