# Df, Sr 데이터들의 연결 실습
- 행 방향 / 열 방향 데이터 연결하기
    - pandas.concat([data,data, ...], axis = 0, join= 'outer')
    - 기본값: 행방향, 합집합 -> 없는 컬럼 데이터는 NaN으로 채우기

    - pandas.DataFrame.append(df)  -> 조만간 없어질 예정(deprecated)
    - 행방향만 연결 

```python
# 모듈 로딩
import pandas as pd
import numpy as np

# df 2개 생성
PATH = r'C:\Users\USER\Documents\GitHub\python_exer\DATA_WOO'
FILENAME  = PATH + '/stock valuation.xlsx'
FILENAME2 = PATH + '/stock price.xlsx'
pd.read_excel(FILENAME)
pd.read_excel(FILENAME2)
```

```python
# 기초 df 생성
df1 = pd.DataFrame({'A': [1,2,3,4,5], 'B': [12,23,34,45,56]})
df2 = pd.DataFrame({'B': [1,3,6], 'C':[2,4,6]})
df1
```

```python
df2
```

## 2개 df 연결 하기[기본값]


```python
# [기본] 행방향, 모든 컬럼 합집합
alldf1 = pd.concat([df1,df2])
alldf1
```

```python
# [수정] 행방향, 모든 컬럼 합집합, 기존 인덱스 무시하고 새롭게 인덱스 설정
alldf2 = pd.concat([df1,df2], ignore_index=True)
alldf2
```

```python
# [수정] 행방향, 공통 컬럼 교집합, 기존 인덱시 무시하고 새롭게 인덱스 설정
alldf3 = pd.concat([df1,df2], ignore_index=True, join='inner')
alldf3
```

```python
# [수정] 행방향, 공통 컬럼 교집합, 기존 인덱시 무시하고 새롭게 인덱스 설정
alldf3 = pd.concat([df1,df2], ignore_index=True, join='inner')
alldf3
```

```python
# [수정] 열방향, 공통 인덱스 합집합
alldf4 = pd.concat([df1,df2], axis=1)
alldf4
```

```python
# [수정] 열방향, 공통 인덱스 합집합, 기존 컬럼명 무시하고 새롭게 컬럼명 설정
alldf5 = pd.concat([df1,df2], axis=1, ignore_index=True)
alldf5
```

```python
# [수정] 열방향, 공통 인덱스 교집합, 기존 컬럼명 무시하고 새롭게 컬럼명 설정, 
alldf6 = pd.concat([df1,df2], axis=1, ignore_index=True, join='inner')
alldf6
```
