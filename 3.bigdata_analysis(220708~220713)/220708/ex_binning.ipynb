# 데이터 정규화
- 수치데이터 컬럼마다 값의 범위가 다름
- 값의 범위: 0 ~ 1, -1 ~ 1로 값을 설정
    - 방법1: 컬럼의 최대값의 절대값으로 전체 데이터를 나누기
    - 방법2: (컬럼의데이터 - 최소값) / (최대값 - 최소값)
    - 방법3: (컬럼의 데이터- 평균)/ 표준편차 -> Z-score

```python
# 방법1: 컬럼의 최대값의 절대값으로 전체 데이터를 나누기
horsedf.insert(horsedf.columns.get_loc('weight')+1, 'weight_normalize', 0)
maximal = abs(horsedf['weight'].max())
horsedf['weight_normalize'] = horsedf['weight'] / maximal
horsedf.head(5)
```

```python
# 방법2: (컬럼의데이터 - 최소값) / (최대값 - 최소값)
minimal = abs(horsedf['weight'].min())
horsedf.insert(horsedf.columns.get_loc('weight_normalize')+1, 'weight_normalize2', 0)
horsedf['weight_normalize2'] = (horsedf['weight']- minimal) /(maximal - minimal)
horsedf.head(5)
```

```python
# 방법3: (컬럼의 데이터- 평균)/ 표준편차 -> Z-score
weight_mean = horsedf['weight'].mean()
weight_std = horsedf['weight'].std()
horsedf.insert(horsedf.columns.get_loc('weight_normalize2')+1, 'weight_normalize3', 0)
horsedf['weight_normalize3'] = (horsedf['weight'] - weight_mean) / weight_std
horsedf.head(5)
```

```python
def normalize(df, col):
    """
    df = DataFrame, col = column(int or float) -> Add normalized_col 
    방식 : (컬럼의 데이터- 평균)/ 표준편차 -> Z-score 
    """
    col_mean = df[col].mean()
    col_std = df[col].std()
    col_nbr = df.columns.get_loc(col)+1
    df.insert(col_nbr, f'{col}_normal', 0)
    df[f'{col}_normal'] = (df[col] - col_mean) / col_std

def catch_int_col(df, sometype):
    """
    df = DataFrame, sometype = type명을 str으로 써주세요 -> return: column 명
    """
    # enumerate 써서 보강할 수 있을듯
    somelist = [(df.dtypes[num] == sometype) for num in range(len(df.columns))]
    return df.loc[:,somelist].columns


someList = catch_int_col(horsedf, 'float64')
for col in someList:
    normalize(horsedf, col)

someList = catch_int_col(horsedf, 'int64')
for col in someList:
    normalize(horsedf, col)
horsedf.head(10)

```
