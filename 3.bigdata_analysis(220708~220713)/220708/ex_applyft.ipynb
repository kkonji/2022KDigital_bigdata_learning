### 사용자 정의 함수 적용 apply() 
    - apply(함수명, axis=방향/축)
    - 원소/요소에 함수 적용
    - 입력: Series의 요소 또는 DataFrame의 시리즈, 즉 행 또는 열
    - 

```python
import numpy as np
# 기본: axis == 0, row의 컬럼마다 함수 적용
subdf.apply(np.sum)
```

```python
# 기본: axis = 1, 컬럼의 row마다 함수 적용
subdf.apply(np.sum, axis=1)
```

```python
# 사용자 정의 함수

def addvalue(x):
    print(f'debug:    {x}\n' ) 
    return x + x
    
def addTwovalue(x,y):
    print(f'debug:{x}, {y}\n') 
    return x+y


subdf.apply(addvalue)
```

```python
subdf.apply(addvalue, axis=1)
```

```python
# series.apply() --> 원소 단위로 들어간다. (axis=1조건 필요업음)
subdf['Age'].apply(addvalue)
```

```python
subdf.apply(addTwovalue, y= 2000)
```

```python
subdf.apply(addTwovalue, y= 154444444444444444444444444444444)
```

```python
# 사용자 정의 함수
def missingValueCheck(sr):
    return sr.isnull()
def missingValueCount(sr):
    return sr.isnull().sum()

subdf.apply(missingValueCheck)
subdf.apply(missingValueCount)
subdf.apply(missingValueCheck, axis=1)
subdf.apply(missingValueCount, axis=1)
```
