# diabetes 데이터셋 관련 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.layers import Dense, Dropout, Flatten
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# 모델 도식화 가능
from keras.utils import plot_model
from keras import Sequential
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

```python
df.info()
```

```python
data = df.iloc[:,:8].to_numpy()
target = df.iloc[:,8].to_numpy()
```

```python
import seaborn as sns
df.corr()
```

```python
# 표준화
st = StandardScaler()
st.fit(data)
data = st.transform(data)
```

```python
print(data.shape, target.shape)
```

```python
train_data, test_data, train_target, test_target = train_test_split(data, target, stratify=target, test_size=0.2, random_state=42)
```

```python
md = Sequential()
md.add(Dense(100, activation='relu', input_shape= (8,)))
md.add(Dropout(0.3))
for i in range(10):
    md.add(Dense(1000, activation='relu'))
    md.add(Dropout(0.3))
md.add(Dense(2, activation='sigmoid'))
md.summary()
```

```python
sgd = tf.keras.optimizers.SGD(momentum=0.9, nesterov=True)
md.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics='accuracy')
md.fit(train_data,train_target, epochs=100, batch_size=40)
md.evaluate(test_data, test_target)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth=7)
rf.fit(train_data, train_target)
print(rf.score(train_data, train_target))
print(rf.score(test_data, test_target))
```
