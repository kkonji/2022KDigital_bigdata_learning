## 모델 저장 실습

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.datasets import load_iris
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from sklearn.model_selection import train_test_split
# 모델 도식화 가능
from sklearn.preprocessing import StandardScaler
from keras.utils import plot_model
from keras import Sequential
from keras.utils import set_random_seed
```

### 데이터 로딩

```python
data = load_iris()['data']
target = load_iris()['target']
target_name = load_iris()['target_names']

train_data, test_data, train_target, test_target = train_test_split(data, target, stratify=target, test_size=0.3, random_state=42)
```

```python
print(train_data.shape, test_data.shape, target_name)
set_random_seed(42)
```

```python
st = StandardScaler()
st.fit(train_data)
train_data = st.transform(train_data)
test_data = st.transform(test_data)
```

```python
md = Sequential()
md.add(Dense(400, activation='relu', input_shape=(4,) ))
md.add(Dropout(0.3))
md.add(Dense(400, activation='relu'))
md.add(Dropout(0.3))
md.add(Dense(400, activation='relu'))
md.add(Dropout(0.3))
md.add(Dense(3, activation='softmax'))
md.summary()
```

```python
sgd = tf.keras.optimizers.SGD(momentum=0.9, nesterov=True)
md.compile(optimizer=sgd, loss='sparse_categorical_crossentropy', metrics='accuracy')
md.fit(train_data, train_target, batch_size=32, epochs=1000)
```

```python
md.evaluate(test_data, test_target)
```

```python
M_FILE = 'irisModel.h5'
from keras.models import save_model
save_model(md, M_FILE)
```

```python
W_FILE = 'irisWeight.h5'
md.save_weights(W_FILE)
```

### 복원 
- 모델 또는 가중치 복원
- load_model (모델 파일 또는 모델 폴더명)
- Sequential.load_weights (가중치 파일 또는 폴더명)

```python
from keras.models import load_model
rmodel = load_model(M_FILE)
rmodel.evaluate(test_data, test_target)
```

### 가중치 복원

```python
# 모델 구조
md2 = Sequential(name='newmodel')
md2.add(Dense(400, activation='relu', input_shape=(4,) ))
md2.add(Dropout(0.3))
md2.add(Dense(400, activation='relu'))
md2.add(Dropout(0.3))
md2.add(Dense(400, activation='relu'))
md2.add(Dropout(0.3))
md2.add(Dense(3, activation='softmax'))
md2.summary()
```

```python
md2.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics='accuracy')
md2.load_weights(W_FILE)
```
