## 정렬
1. 카운팅 정렬 : 배열에 존재하는 수의 개수를 세고, 이 정보를 바탕으로 정렬을 수행
    - 수와 수의 개수는 배열 or 딕셔너리의 인덱스와 해당 값을 통해서 기록
    - O(n+k), k: 배열의 최댓값, k < n일때 효과가 좋고, k >= n이면 성능이 좋지 않다.
2. 

```python
arr = [50, 24, 10, 60, 30, 70, 35, 1, 24, 4, 7, 8, 8 , 9, 16, 16, 25, 85, 99, 3, 6]
count = [0] * 101
for num in arr:
    count[num] += 1

for i in range(1, len(count)):
    count[i] += count[i-1]

result = [0] * len(arr)
for num in arr:
    idx = count[num]
    result[idx-1] = num
    count[num] -= 1
    print(result)
print(result)
```
