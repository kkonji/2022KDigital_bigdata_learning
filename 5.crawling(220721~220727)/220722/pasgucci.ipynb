```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
def taker(line, num):
    """
    param: List or Series or Array -> return: list
    List 또는 Series 또는 Array를 받아와서 인덱스 기준으로 num줄씩 잘라서 리스트로 내보내는 함수 
    """
    somelist = []
    for i in range(len(line)// num):
        somelist.append(line[num*i:num*(i+1)].to_list())
    return somelist
```

```python
elements = []
for num in range(1, 55):
    url = f'https://www.caffe-pascucci.co.kr/store/storeList.asp?page={num}'
    html = requests.get(url)
    soup = BeautifulSoup(html.text, 'html.parser')
    # 0: 번호 1: 가게명 2: 지역 3:부가정보 4: 주소가 들어있음 5: 전화번호 6: 상세지도보기

    somelist = soup.select('table.storeTable td')
    series = pd.Series(somelist).apply(lambda z: z.text)
    series = taker(series, 7)
    for value in series:
        elements.append(value)
print(elements)
df = pd.DataFrame(elements)
```

```python
df = pd.DataFrame(elements, columns=['번호', '상호명','지역', '상세정보', '주소', '전화번호', '상세지도' ])
new_df = df[['상호명', '지역', '주소', '전화번호']]
new_df['주소'] = new_df['주소'].str.strip().str.split('\n', expand=True)[0]
```

```python
new_df['전화번호'] = new_df['전화번호'].str.strip()
new_df
```

```python
new_df['전화번호'] = new_df['전화번호'].str.replace('.', '-')
new_df
```
