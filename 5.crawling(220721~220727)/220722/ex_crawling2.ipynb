### 단일 도메인 내의 이동
- 필요한 정보만 가져오기 위해, 위키백과 페이지를 분석할 필요
- div 태그의 id='bodycontent' 내부에 있음
- url에는 콜론이 포함되어있지 않음
- url은 /wiki/로 시작

```python
url = 'https://en.wikipedia.org/wiki/Kevin_Bacon'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')

# div 태그의 id='bodycontent' 내부로 들어감
body_content = soup.select_one('div#bodyContent')
body_content

#전방 부정 탐색
pattern = '^(/wiki/)((?!:).)*$' # url에는 콜론이 포함되어있지 않음 + url은 /wiki/로 시작
for link in body_content.find_all('a', href=re.compile(pattern)):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

### 링크간 무작위 이동하기: 동작 과정
- getlinks(articleUrl) 함수작성
- 파라미터: 임의의 url링크를 받음
- 리턴값: 해당링크의 모든 url목록을 리턴(리스트 형태)

```python
import datetime
import random
random.seed(datetime.datetime.now())

def getlinks(articleUrl):
    html = requests.get(f'https://en.wikipedia.org{articleUrl}')
    soup = BeautifulSoup(html.text, 'html.parser')
    pattern = '^(/wiki/)((?!:).)*$'
    return soup.find('div', id='bodyContent').find_all('a', href = re.compile(pattern))

links = getlinks('/wiki/kevin_Bacon')
while (len(links)) >0 :
    num = random.randint(0, len(links)-1 )
    newArticle = links[num].attrs['href']
    print(newArticle)
    links = getlinks(newArticle)
```

### 같은 페이지를 두 번이상 크롤링하지 않기
- set을 사용

```python
pages = set()
def getLinks(url):
    global pages
    html = requests.get(f'https://en.wikipedia.org{url}')
    soup = BeautifulSoup(html.text, 'html.parser')
    pattern = '^(/wiki/)((?!:).)*$'
    for link in soup.find_all('a',href = re.compile(pattern)):
        if link.has_attr('href'):
            if link.attrs['href'] not in pages:
                newpage = link.attrs['href']
                print(newpage)
                pages.add(newpage)
                getLinks(newpage)
getLinks('/wiki/kevin_Bacon')
```
