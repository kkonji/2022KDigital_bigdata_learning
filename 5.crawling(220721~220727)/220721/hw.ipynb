```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.Ytjsz3ZBxPY'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')
```

```python
def des_print(period_name, short_desc, temp, img):
    for i in range(9):
        # image = img[i]['title']
        print('------------------------------------------------------------------------------')
        print(f'[Period]: {period_name[i]}') 
        print(f'[Short desc]: {short_desc[i]}') 
        print(f'[Temperature]: {temp[i]}') 
        print(f'[Image desc]: {img[i]}') 
```

```python
def scraping_use_find(html):
    
    soup = BeautifulSoup(html.text, 'html.parser')
    soup1 = soup.find('div', {'id' : "seven-day-forecast"})

    period_name = pd.Series(soup1.find_all('p', {'class': 'period-name'} )).apply(lambda x: x.text)
    short_desc = pd.Series(soup1.find_all('p',{'class': 'short-desc'} )).apply(lambda x: x.text)
    img = pd.Series(soup1.find_all('img' )).apply(lambda x: x['title'])
    temp = pd.Series(soup1.find_all('p', {'class': 'temp'} )).apply(lambda x: x.text)

    print('[select 함수 이용]')
    des_print(period_name, short_desc, temp, img)
    

scraping_use_find(html)
```

```python
def scraping_use_select(html):
    soup = BeautifulSoup(html.text, 'html.parser')
    period_name = soup.select('#seven-day-forecast .period-name')  # period-name 리스트 형태로 저장 .text 만 붙이기
    short_desc = soup.select('#seven-day-forecast .short-desc')  # ''
    img = soup.select('#seven-day-forecast img')                      # [원소이름]['alt']
    temp = soup.select('#seven-day-forecast .temp')              # ''
    print('[select 함수 이용]')
    for i in range(9):
        print('------------------------------------------------------------------------------')
        print(f'[Period]: {period_name[i].text}') 
        print(f'[Short desc]: {short_desc[i].text}') 
        print(f'[Temperature]: {temp[i].text}') 
        print(f'[Image desc]: {img[i]["title"]}') 


scraping_use_select(html)
```
