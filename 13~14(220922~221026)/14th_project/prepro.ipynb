### 적조 데이터 전처리
- 목표: dataset 통합해서 적조여부 출력
- 출처 : 적조관리 시스템 https://www.nifs.go.kr/rtm/TRS/gispop/redtide.jsp

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
%pip install xlrd
```

```python
pd.read_excel("./data/19_redtide.xls")
```

```python
PATH = "./data/"
df = pd.DataFrame(columns=["예찰일자", "발생해역", "적조생물", "밀도(개체수/mL)", "수온(℃)" , "염분", "좌표"])
for i in range(17,23):
    df = pd.concat([df, pd.read_excel(PATH+f"{i}_redtide.xls")])
```

```python
df = df.drop(columns=["수온(℃)", "염분", "좌표"])
df
```

### 적조생물 값 확인

```python
df["적조생물"].unique()
```

### 적조생물 컬럼에서 불필요한 값에 해당되는 행 64개 제거 -> 728개에서 664개

```python
df = df.loc[~df["적조생물"].isin([' ', '없음', '미 확인']),:]
df
```

```python
df["적조생물"].value_counts()
```

### 거제 또는 통영에 해당되는 행

```python
df.loc[df["발생해역"].str.contains("거제|통영", regex=True),:]
```

### 거제와 통영 모두 해당되는 행

```python
df.loc[(df["발생해역"].str.contains("거제", regex=True)) & (df["발생해역"].str.contains("통영", regex=True)),:]
```

### NaN 값 개수

```python
df.isna().sum()
```

### 밀도 컬럼의 결측값 제거, 데이터 개수: 664 -> 571

```python
df = df.loc[~df["밀도(개체수/mL)"].isna(),:]
df
```
