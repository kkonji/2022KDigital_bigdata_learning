## DataFrame 적조밀도 마지막 전처리

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("redtide.csv")
df
```

### 거제 또는 통영에 해당되는 행

```python
df = df.loc[df["발생해역"].str.contains("거제|통영", regex=True),:]
df
```

```python
df.loc[(df["발생해역"].str.contains("거제", regex=True)) & (df["발생해역"].str.contains("통영", regex=True)),:].index
```

```python
b = [["2019-09-13","거제","Cochlodinium polykrikoides", "512~1440"], ["2019-09-13","통영","Cochlodinium polykrikoides", "512~1440"],
["2019-09-12", "거제", "Cochlodinium polykrikoides", "520~1640"], ["2019-09-12", "통영", "Cochlodinium polykrikoides", "520~1640"],
["2019-09-09", "거제", "Cochlodinium polykrikoides", "592~1500"], ["2019-09-09", "통영", "Cochlodinium polykrikoides", "592~1500"]]
df.drop(df.loc[(df["발생해역"].str.contains("거제", regex=True)) & (df["발생해역"].str.contains("통영", regex=True)),:].index, inplace=True)
```

```python
df = pd.concat([df,pd.DataFrame(b, columns=df.columns)])
df
```

```python
df.reset_index(inplace=True)
df.drop(columns=["index"], inplace=True)
```

```python
df.loc[df.loc[df["발생해역"].str.contains("거제", regex=True),:].index, "발생해역"] = "거제"
```

```python
df.loc[df.loc[df["발생해역"].str.contains("통영", regex=True),:].index, "발생해역"] = "통영"
```

```python
dg = df["밀도(개체수/mL)"].str.split("~", expand=True, )
dg
```

```python
dg = dg.replace("", 0)
```

```python
dg.loc[dg.loc[:,1] == 0,1] = [5000,350]
```

```python
df.loc[:,"밀도(개체수/mL)"] = dg.loc[:,1]
```

```python
df
```

```python
df = df.sort_values(["예찰일자"])
df.to_csv("redtide_통영_거제.csv", index=0)
```
